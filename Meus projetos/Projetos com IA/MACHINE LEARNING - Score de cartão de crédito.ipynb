```python
import pandas as pd

# Step 1: Importing the database
# Passo 1: Importando a base de dados

# Importing the table
# Importando a tablea
tabela = pd.read_csv('clientes.csv')

# Deleting the column 'id_cliente', because it will be useless
# Excluindo a coluna 'id_cliente', já que ela será inútil
tabela = tabela.drop(columns='id_cliente')

# Showing the table
# Mostrando a tabela
display(tabela)
display(tabela.info())
```

```python
# Step 2: Preparing the database for the AI
# Passo 2: Preparar a base de dados para a inteligência artificial
# Label Encoder - profissao, mix_credito, comportamento_pagamento

# profissao
# engenheiro -> 1
# cientista -> 2
# advogado -> 3
# artista -> 4
# mecanico -> 5

# Importing the label encoder
# Importar o label encoder
from sklearn.preprocessing import LabelEncoder

# Changing the name of the label encoder to codificador, to be easier
# Mudando o nome do label encoder para codificador, para ficar mais fácil
codificador = LabelEncoder()

# Aplying the label encoder on the columns
# Aplicar o label encoder nas colunas

# profissao
tabela['profissao'] = codificador.fit_transform(tabela['profissao'])

# mix_credito
tabela['mix_credito'] = codificador.fit_transform(tabela['mix_credito'])

# comportamento_pagamento
tabela['comportamento_pagamento'] = codificador.fit_transform(tabela['comportamento_pagamento'])
display(tabela.info())
```

```python
# Who I wanna predict (y), who I wanna use to do the predict (x)
# quem eu quero prever (y), quem eu quero usar pra fazer a previsão (x)
y = tabela['score_credito']
x = tabela.drop(columns='score_credito')

# training and testing
# treino e teste
from sklearn.model_selection import train_test_split


'''Doing the subdivisions of the variables and showing to the AI which of them are going to train and which of them are going to test
x_treino = Subset of input data for training.
x_teste = Subset of input data for testing.
y_treino = Subset of labels or target values correspondents to x_treino
y_teste = Subset of labels or target values correspondents to x_teste
30% of all database are going to be used for testing the AI, 70% are going to be used for training the AI'''


''' fazendo as subdivisões das variáveis e mostrando para a IA quais delas vão ser para treinar e quais delas vão ser para testar
x_treino: Subconjunto de dados de entrada para treinamento.
x_teste: Subconjunto de dados de entrada para teste.
y_treino: Subconjunto de rótulos ou valores alvo correspondentes ao x_treino.
y_teste: Subconjunto de rótulos ou valores alvo correspondentes ao x_teste.
30% de toda a base de dados vai servir para testar a IA, 70% vai ser para treinar ela'''

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# creating the AI
# criar a inteligência artificial
# Decision Tree
# Árvore de Decisão
# KNN -> Vizinhos próximos (Nearest Neighbors)

# import the AI
# importa a inteligência artificial
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# create the AI
# cria a iteligêcia artificial
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# train the AI
# treina a inteligência artificial
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

# test the models
# testar os modelos
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_arvoredecisao.predict(x_teste)

from sklearn.metrics import accuracy_score

# testing what is the best model
# testando qual é o melhor modelo
print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```

```python
# the 2 models are the same in precision, so I can choose what it's gonna be better for me, so I gonna choose the model Decision Tree
# os 2 modelos são iguais em questão de precisão, então eu posso escolher o que funcionar melhor para mim, então eu vou escolher o modelo Árvore de Decisão
# Pick another database with the new clients
# Pego uma database só com novos clientes
tabela_nova = pd.read_csv('novos_clientes.csv')
display(tabela_nova)

# Do the Label Encoder on every column of the new table
# Faço o Label Encoder em cada coluna da tabela nova

# profissão
tabela_nova['profissao'] = codificador.fit_transform(tabela_nova['profissao'])

# mix_credito
tabela_nova['mix_credito'] = codificador.fit_transform(tabela_nova['mix_credito'])

# comportamento_pagamento
tabela_nova['comportamento_pagamento'] = codificador.fit_transform(tabela_nova['comportamento_pagamento'])

# Do the predict on the new table
# Faz a previsão da tabela nova
# from now on, every new client will entry into this table and it's just execute this software that AI predict if the client's score is good, poor or standard
# Agora em diante, cada cliente novo vai entrar nessa tabela e é só executar esse software que a IA vai dizer se o score do cliente é bom, ruim ou mais ou menos
# Poor = Ruim       Standard = Mais ou menos          Good = Bom
previsoes = modelo_arvoredecisao.predict(tabela_nova)
display(previsoes)
```
